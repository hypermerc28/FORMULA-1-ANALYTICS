```python
#Question4-Calculate the historical probability of each country of having a driver in the podium?

# To calculate the historical probability of each country having a driver on the podium, we need to:
# 1. Count the number of times drivers from each country have finished in the top 3 (podium).
# 2. Divide this count by the total number of race participations for each country.

# Counting the number of podium finishes for each country
podium_finishes_by_country = F1[F1["final_position"].isin([1, 2, 3])].groupby("driver_nationality")["final_position"].count()

# Counting race participations for each driver nationality 
participations_by_nationality = F1["driver_nationality"].value_counts()

# Dividing the count of podium finishes by the total number of race participations for each country
probability_of_podium_by_country = (podium_finishes_by_country / participations_by_nationality)*100

probability_of_podium_by_country.sort_values(ascending=False)
```

```python
#Question4-Calculate the historical probability of each country of having a driver win a race

# To calculate the historical probability of each country having a driver win a race, we need to:
# 1. Count the number of times drivers from each country have won a race.
# 2. Divide this count by the total number of race participations for each country.

# Counting the number of race wins for each country
race_wins_by_country = F1[F1["final_position"] == 1].groupby("driver_nationality")["final_position"].count()

# Counting race participations for each driver nationality 
participations_by_nationality = F1["driver_nationality"].value_counts()

# Calculating the probability of a win for each country
probability_of_win_by_country = (race_wins_by_country / participations_by_nationality) * 100

probability_of_win_by_country.sort_values(ascending=False)
```

### 5. Circuit analytics (2 points)

* Which would you say is the toughest circuit in Formula 1?  (0.5 point)
* Which circuit requires the most f1 experience to win? (0.5 point)
* Which circuit and year saw the most number of non-finishers? (0.5 point)
* For each constructor, which is their best circuit in terms of amount of podiums? (0.5 point)

```python
#Question5-Which would you say is the toughest circuit in Formula 1? First Approach
# We calculateed the average DNF (Did Not Finish) ratio for each circuit in Formula 1, we first identified the total number of races and the number of DNFs for each circuit. 
#The DNF ratio can be calculated by dividing the number of DNFs by the total number of races at each circuit

# Identifying DNFs
# In F1, a driver is classified as DNF if their final position is NaN or if the status is not 'Finished'
F1["DNF"] = F1["final_position"].isna() | (F1["status"] != "Finished")

# Grouping by circuit and calculating the DNF ratio
circuit_dnf_data = F1.groupby("circuit_name").agg(
    total_races=("DNF", "count"),
    total_dnf=("DNF", "sum")
)
circuit_dnf_data["dnf_ratio"] = (circuit_dnf_data["total_dnf"] / circuit_dnf_data["total_races"]) * 100

# Sorting circuits by highest DNF ratio
toughest_circuits = circuit_dnf_data.sort_values(by="dnf_ratio", ascending=False)

toughest_circuits.head()
```

```python
#Question5-Which would you say is the toughest circuit in Formula 1? Second Approach

# To find the circuit where many drivers could not finish the race  we will:
# 1. Count the total number of DNF (Did Not Finish) for each circuit.
# 2. Identify the circuit with the highest number of DNFs.

# Creating a column 'dnf' indicating whether the race ended in Did Not Finish
F1["dnf"] = F1["status"] != "Finished"

# Counting the total number of DNFs for each circuit
total_dnf_by_circuit = F1[F1["dnf"]].groupby("circuit_name")["dnf"].count()

# Identifying the circuit with the highest number of DNFs
circuit_most_dnf = total_dnf_by_circuit.idxmax()
most_dnf_count = total_dnf_by_circuit.max()

circuit_most_dnf, most_dnf_count

```

```python
#Question5-Which circuit requires the most f1 experience to win? 

# Calculating the average experience of winners for each circuit and identifying the one with the highest average experience

# To determine which circuit requires the most F1 experience to win, we can calculate the average experience of winners at each circuit.
# Experience can be measured as the number of years since a driver's first race until they win at each circuit.

# Calculating the year of each driver's first race
first_race_year = F1.groupby("driver")["year"].min()

# Merging the first race year with the main dataset
f1_data_with_experience = F1.merge(first_race_year.rename("first_race_year"), on="driver")

# Calculating the experience (in years) of drivers at the time of each race
f1_data_with_experience["experience_at_race"] = f1_data_with_experience["year"] - f1_data_with_experience["first_race_year"]

# Filtering for races that were won
winning_races = f1_data_with_experience[f1_data_with_experience["final_position"] == 1]

# Calculating the average experience of winners at each circuit
average_experience_of_winners = winning_races.groupby("circuit_name")["experience_at_race"].mean()

# Identifying the circuit that requires the most experience to win
circuit_most_experience = average_experience_of_winners.idxmax()
highest_average_experience = average_experience_of_winners.max()

circuit_most_experience, highest_average_experience


```

```python
#Question5-Which circuit and year saw the most number of non-finishers?

# To find the circuit and year with the most non-finishers (where 'status' is not equal to 'Finished'), we will count these instances

# Filtering data for non-finishers (where 'status' is not 'Finished')
non_finishers = F1[F1["status"] != "Finished"]

# Counting the number of non-finishers for each circuit in each year
non_finishers_count = non_finishers.groupby(["year", "circuit_name"]).size()

# Finding the circuit and year with the most non-finishers
most_non_finishers = non_finishers_count.idxmax()
most_non_finishers_count = non_finishers_count.max()

most_non_finishers, most_non_finishers_count

```

```python
#Question5-For each constructor, which is their best circuit in terms of amount of podiums?


# Filter for podium finishes
podium_finishes = F1[F1["final_position"].isin([1, 2, 3])]

# Group by constructor and circuit, count podiums
podiums_by_constructor = podium_finishes.groupby(["constructor_name", "circuit_name"]).size()

# Reset index and sort
podiums_by_constructor = podiums_by_constructor.reset_index(name="podium_count")
podiums_by_constructor = podiums_by_constructor.sort_values(["constructor_name", "podium_count"], ascending=[True, False])

# Find the best circuit for each constructor along with the podium count
best_circuit_for_each_constructor = podiums_by_constructor.groupby("constructor_name").first()

# Sorting by podium count in descending order
sorted_best_circuits = best_circuit_for_each_constructor.sort_values("podium_count", ascending=False)

sorted_best_circuits

```

```python
# Defining a function to convert lap times to seconds for comparison
def lap_time_to_seconds(lap_time):
    time_parts = lap_time.split(':')
    minutes = int(time_parts[0])
    seconds = float(time_parts[1])
    return 60 * minutes + seconds

# Filtering data for Ferrari's fastest laps with valid fastest lap time
ferrari_fastest_lap_times_data = F1[(F1['constructor_name'] == 'Ferrari') & (~F1['fastest_lap_time'].isna())]

# Converting lap times to seconds
ferrari_fastest_lap_times_data['fastest_lap_time_seconds'] = ferrari_fastest_lap_times_data['fastest_lap_time'].apply(lap_time_to_seconds)

# Grouping the fastest lap times by circuit for Ferrari and finding the minimum (fastest) lap time
ferrari_fastest_lap_time_by_circuit = ferrari_fastest_lap_times_data.groupby(['circuit_name'])['fastest_lap_time_seconds'].min().reset_index()

# Creating a horizontal bar chart for the fastest lap time per circuit
fig_ferrari_fastest_lap_time_circuit = px.bar(ferrari_fastest_lap_time_by_circuit, y='circuit_name', x='fastest_lap_time_seconds',
                                              labels={'fastest_lap_time_seconds': 'Fastest Lap Time (Seconds)', 'circuit_name': 'Circuit'},
                                              title="Fastest Lap Time per Circuit for Ferrari Team",
                                              orientation='h')

# Applying the specified color and font settings
fig_ferrari_fastest_lap_time_circuit.update_layout(
    plot_bgcolor='rgba(0,0,0,0)',
    paper_bgcolor='rgba(0,0,0,0)',
    font=dict(
        family="Courier New, monospace",
        size=8,
        color="#7f7f7f"
    ),
    yaxis={'categoryorder':'total ascending'}
)

fig_ferrari_fastest_lap_time_circuit.show()

```
